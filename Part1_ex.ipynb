```r
library(tidyverse)
data <- read_csv("/content/hypothesis_testing_exercise_data.csv")
head(data)
```

```r
data$treatment
```

```r
x1 <- data[data$treatment == "A", "bp_reduction", drop = TRUE]
x2 <- data[data$treatment == "B", "bp_reduction", drop = TRUE]
```

```r
ggplot(data=NULL) + geom_histogram(aes(x=x1, y=..density..), fill="red", alpha=0.5, bins=20) +
                    geom_histogram(aes(x=x2, y=..density..), fill="blue", alpha=0.5, bins=20) +
                    theme_bw(base_size = 18) + labs(x="Blood Pressure Reduction", y="Density")
```

```r
ggplot(data=data, aes(x=bp_reduction, fill=treatment)) + geom_histogram(col="gray", bins=20) + theme_bw(base_size = 18)
```

```r
ggplot(data=NULL) + geom_histogram(aes(x=x1), fill="red", alpha=0.5, bins=20) +
                    geom_histogram(aes(x=x2), fill="blue", alpha=0.5, bins=20) +
                    theme_bw(base_size = 18) + labs(x="Blood Pressure Reduction", y="Counts")
```

```r
qqnorm(x1, main = "Normal qq-plot (Benign masses)",
       ylab = "Empirical quantiles",
       xlab = "Theoretical quantiles")

qqline(x1, col="red", lty=4, lwd=2)
```

```r
qqnorm(x2, main = "Normal qq-plot (Benign masses)",
       ylab = "Empirical quantiles",
       xlab = "Theoretical quantiles")

qqline(x2, col="red", lty=4, lwd=2)
```

```r
t.test(x1, x2)
```
