```r
data <- data %>% select(-c("...33"))
```

```r
colnames(data)
```

```r
data <- data %>%
  set_names(
    ~ .x %>%
      str_replace_all("_mean", "")
  )
  head(data)
```

```r
radius_M <- data[data$diagnosis=="M", "radius", drop=TRUE]
radius_B <- data[data$diagnosis=="B", "radius", drop=TRUE]

breaks <- seq(min(data$radius), max(data$radius), length.out = 30)
```

```r
hist(radius_B, breaks = breaks, prob = TRUE, col="violetred", xlab = "Radius")
hist(radius_M, breaks = breaks, prob = TRUE, col="#5800d3", add = TRUE)
legend("topright", legend = c("B", "M"), fill = c("violetred", "#5800d3"), border="black")
```
