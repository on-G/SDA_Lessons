```r
t_stat_obs <- t.test(x3, x4)$statistic
print(t_stat_obs)
```

```r
set.seed(101)
nsim <- 1000
t_stat_perm <- c()

for(i in 1:nsim){
  diagnosis_perm <- sample(data$diagnosis, replace = FALSE)
  x1_perm <- data[ diagnosis_perm == "B", "compactness" ]
  x2_perm <- data[ diagnosis_perm == "M", "compactness" ]

  t_stat_perm[i] <- t.test(x1_perm, x2_perm)$statistic
}
```

```r
ggplot(data = NULL, aes(x = t_stat_perm)) + geom_histogram(bins=100, fill="#7474d1", color="black") +
                                            geom_vline(xintercept = t_stat_obs, color = "red", linewidth = 1, lty = 4) +
                                            xlab("t statistic") + theme_bw(base_size = 18)
```

```r
p_value <- 2 * sum(t_stat_perm < t_stat_obs) / length(t_stat_perm)
print(p_value)
```

```r
feature_list <- colnames(data)
feature_list <- feature_list[!feature_list %in% c("id", "diagnosis")]

p_values <- c()

for(feature_i in feature_list){
  x1_i <- data[data$diagnosis == "B", feature_i, drop = TRUE]
  x2_i <- data[data$diagnosis == "M", feature_i, drop = TRUE]

  t_test_result_i <- t.test(x1_i, x2_i)

  p_values <- c(p_values, t_test_result_i$p.value)
}
```

```r
results_df <- data.frame(Feature=feature_list, PValue=p_values)
results_df
```
